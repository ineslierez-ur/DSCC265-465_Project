```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('X_same_notnormalized.csv')
df
```

```python
df = df.drop(["Unnamed: 0"], axis=1)
pd.set_option('display.max_columns', None)
df.head()
```

```python
X = df
y = pd.read_csv('y.csv')
y = y.drop(["Unnamed: 0"], axis=1)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, 
                            test_size=.3, random_state=265)
```

# Ridge Regression

```python
from sklearn.linear_model import Ridge
from sklearn.metrics import mean_absolute_error, r2_score

ridge = Ridge(alpha=0.1)
_ = ridge.fit(X_train, y_train)
preds = ridge.predict(X_test)

print("Training score:", ridge.score(X_train, y_train))
print("Testing score:", ridge.score(X_test, y_test))
print("MAE of Ridge Regression:", mean_absolute_error(y_test, preds))
print("R2 of Ridge Regression:", r2_score(y_test,preds),'\n')
```

```python
from sklearn.linear_model import RidgeCV

ridge = RidgeCV(alphas=np.arange(1, 100, 5), scoring='r2', cv=10)
_ = ridge.fit(X, y)
ridge.alpha_
```

```python
ridge = Ridge(alpha=6)
_ = ridge.fit(X_train, y_train)
preds = ridge.predict(X_test)

print("Training score:", ridge.score(X_train, y_train))
print("Testing score:", ridge.score(X_test, y_test))
print("MAE of Ridge Regression CV:", mean_absolute_error(y_test, preds))
print("R2 of Ridge Regression CV:", r2_score(y_test,preds),'\n')
```

# Lasso Regression

```python
from sklearn.linear_model import Lasso, LassoCV

lasso = LassoCV(alphas=np.arange(0.000000001, 1, 0.05), cv=10)
_ = lasso.fit(X, y)

print('Best alpha:', lasso.alpha_)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, 
                  test_size=.3, random_state=1121218)
lasso = Lasso(alpha=1e-09)
lasso.fit(X_train, y_train)
preds = lasso.predict(X_test)

print("Training score:", lasso.score(X_train, y_train))
print("Testing score:", lasso.score(X_test, y_test))
print("MAE of Lasso Regression:", mean_absolute_error(y_test, preds))
print("R2 of Lasso Regression:", r2_score(y_test,preds),'\n')
```

# ElasticNet

```python
from sklearn.linear_model import ElasticNet, ElasticNetCV

enet_model = ElasticNet().fit(X_train, y_train)
enet_cv_model = ElasticNetCV(cv = 10).fit(X_train,y_train)
```

```python
from sklearn.metrics import mean_squared_error,r2_score
enet_tuned = ElasticNet(alpha = enet_cv_model.alpha_).fit(X_train,y_train)

preds = enet_tuned.predict(X_test)
print("Training score:", enet_tuned.score(X_train, y_train))
print("Testing score:", enet_tuned.score(X_test, y_test))
print("MAE of Lasso Regression:", mean_absolute_error(y_test, preds))
print("R2 of ElasticNet Regression:", r2_score(y_test,preds),'\n')
```

# Random Forest Model

```python
# Random Forest Model
https://towardsdatascience.com/hyperparameter-tuning-the-random-forest-in-python-using-scikit-learn-28d2aa77dd74
```

```python
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor(random_state = 42)
from pprint import pprint

from sklearn.model_selection import RandomizedSearchCV
# Number of trees in random forest
n_estimators = [int(x) for x in np.linspace(start = 200, stop = 2000, num = 10)]
# Number of features to consider at every split
max_features = ['auto', 'sqrt']
# Maximum number of levels in tree
max_depth = [int(x) for x in np.linspace(10, 110, num = 11)]
max_depth.append(None)
# Minimum number of samples required to split a node
min_samples_split = [2, 5, 10]
# Minimum number of samples required at each leaf node
min_samples_leaf = [1, 2, 4]
# Method of selecting samples for training each tree
bootstrap = [True, False]
# Create the random grid
random_grid = {'n_estimators': n_estimators,
               'max_features': max_features,
               'max_depth': max_depth,
               'min_samples_split': min_samples_split,
               'min_samples_leaf': min_samples_leaf,
               'bootstrap': bootstrap}
pprint(random_grid)
```

```python
# Use the random grid to search for best hyperparameters
# First create the base model to tune
rf = RandomForestRegressor()
# Random search of parameters, using 3 fold cross validation, 
# search across 10 different combinations, and use all available cores
rf_random = RandomizedSearchCV(estimator = rf, param_distributions = random_grid, n_iter = 10, cv = 3, verbose=2, random_state=42, n_jobs = -1)
# Fit the random search model
rf_random.fit(X_train, y_train)
```

```python
rf_random.best_params_
```

```python
from sklearn.metrics import r2_score
def evaluate(model, X_test, y_test):
    predictions = model.predict(X_test)
    predictions = predictions.reshape(predictions.size,1)
    errors = abs(np.subtract(predictions,y_test))
    mape = 100 * np.mean(errors / y_test)[0]
    #mape_accuracy = (100 - mape)[0]
    accuracy = r2_score(predictions,y_test)
    meanErrors = np.mean(errors)[0]
    print('Model Performance')
    #print('Average Error: {:0.3f} degrees.'.format(meanErrors))
    
    print('R2_score = {:0.2f}.'.format(accuracy))
    print('RMSE: {:0.3f}.'.format(mean_squared_error(y_test, predictions, squared=False)))
    print('MAPE = {:0.1f}%.'.format(mape))
    
    return accuracy
base_model = RandomForestRegressor(n_estimators = 10, random_state = 42)
base_model.fit(X_train, y_train)
base_accuracy = evaluate(base_model, X_test, y_test)
```

```python
best_random = rf_random.best_estimator_
random_accuracy = evaluate(best_random, X_test, y_test)
```

```python
print('Improvement of {:0.2f}.'.format( 100 * (random_accuracy - base_accuracy) / base_accuracy))
```

# Support Vector Regressor 
http://scikit-learn.sourceforge.net/0.5/auto_examples/svm/plot_svm_regression.html

```python
from sklearn.svm import SVR
svr_rbf = SVR(kernel='rbf', C=1e4, gamma=0.1)
y_rbf = svr_rbf.fit(X_train, y_train).predict(X_test)
```

```python
svr_lin = SVR(kernel='linear', C=1e4)
y_lin = svr_lin.fit(X_train, y_train).predict(X_test)
```

```python
svr_poly = SVR(kernel='poly', C=1e4, degree=2)
y_poly = svr_poly.fit(X_train, y_train).predict(X_test)
```

```python
import pylab as pl
pl.scatter(X_test, y_test, c='k', label='data')
pl.hold('on')
pl.plot(X_test, y_rbf, c='g', label='RBF model')
pl.plot(X_test, y_lin, c='r', label='Linear model')
pl.plot(X_test, y_poly, c='b', label='Polynomial model')
pl.xlabel('data')
pl.ylabel('target')
pl.title('Support Vector Regression')
pl.legend()
pl.show()
```
